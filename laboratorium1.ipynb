## Zadanie 3
Sporządzenie wykresu funkcji f(x) = x^2 + 5 dla 3 różnych przedziałów.

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
#definicja funkcji
f = lambda x: x ** 2 + 5
#definicja przedziałów argumentów funkcji
x1 = np.linspace(-1,1)
x2 = np.linspace(-6,6)
x3 = np.linspace(0,5)
```

```python
#przedstawienie wykresów funkcji dla danych argumentów
plt.plot(x1, f(x1))
plt.title("f(x)=x^2 + 5")
plt.xlabel("argument")
plt.ylabel("wartość funkcji")
plt.legend("f(x)=x^2 + 5")
plt.grid()
plt.show()
```

```python
plt.plot(x2, f(x2))
plt.title("f(x)=x^2 + 5")
plt.xlabel("argument")
plt.ylabel("wartość funkcji")
plt.legend("f(x)=x^2 + 5")
plt.grid()
plt.show()
```

```python
plt.plot(x3, f(x3))
plt.title("f(x)=x^2 + 5")
plt.xlabel("argument")
plt.ylabel("wartość funkcji")
plt.legend("f(x)=x^2 + 5")
plt.grid()
plt.show()
```

## Zadanie 4
Utworzenie dataframe, w którym kolumny mają nazwy: name, surname, age, sex oraz uzupełniłem je dowolnymi pięcioma rekordami.

```python
import pandas as pd

d = {'name': ['Jan', 'Domnik', 'Andrzej', 'Michał', 'Szymon'],'surname': ['Kowalski', 'Nowak', 'Janik', 'Hesek', 'Holajn'],'age': [30,45,22,21,21], 'sex': ['male','male','male','male','male']}
dataframe = pd.DataFrame(data=d)
print(dataframe)
```

Informacje o danych

```python
print(dataframe.info())
```

Opis danych

```python
print(dataframe.describe())
```
